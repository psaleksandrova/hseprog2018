```python
import random
COUNT_END = 7


def get_word (filename):
    # открытие определённого файла в зависимости от выбора пользователя
    # выбор рандомного слова из файла для игры
    filename = filename + '.txt'
    wfile = open(filename, 'r', encoding='utf-8')
    words = wfile.read().split('\n')
    word = random.choice(words)
    return word


def attempt_to_str (count_attempt):
    # вывод оставшегося числа попыток, используя нужное склонение
    # функция работает только при конкретном ограничении на максимальное число попыток, равное 7
    if count_attempt == 1:
        return 'У Вас осталась 1 попытка.'
    if 2 <= count_attempt <= 4:
        return 'У Вас осталось %d попытки.' % count_attempt
    return 'У Вас осталось %d попыток.' % count_attempt


def draw_picture (count_attempt):
    # дорисовка новых деталей виселицы
    wfile = open(str(count_attempt) + '.txt', 'r', encoding='utf-8')
    for line in wfile:
        print(line, end = '')
    print('\n')


def apply_letter (letter, word, ans, count_attempt):
    # проверка заданной буквы на наличие в слове
    # преобразование текущей строки из уже угаданных букв в случае удачной попытки
    pos = word.find(letter)
    if pos == -1:
        count_attempt -= 1
        print('К сожалению, такой буквы нет в загаданном слове.')
        draw_picture(count_attempt)
        if count_attempt != 0:
            print(attempt_to_str(count_attempt))
    else:
        print('Такая буква есть в загаданном слове.')
        while pos != -1:
            ans = ans[:pos] + letter + ans[(pos + 1):]
            word = word[:pos] + '_' + word[(pos + 1):]
            pos = word.find(letter)
    return ans, count_attempt


def check_input (letter, dict_letters):
    # проверка буквы, которую воодит пользователь
    if len(letter) > 1 or not (ord('а') <= ord(letter) <= ord('я')):
        # пользователем был введён некорректный символ
        return 1
    if letter not in dict_letters:
        # пользователь ввёл букву, которая уже была использована
        return 2
    return 0
```

```python
def main():
    count_attempt = COUNT_END
    dict_letters = {chr(c) for c in range(ord('а'), ord('я') + 1)}
    
    dict_filenames = {'породы собак', 'напитки', 'грибы'}
    filename = input('Выберите одну из предложенных тем: породы собак, напитки или грибы. ').lower()
    while filename not in dict_filenames:
        filename = input('Извините, надо выбрать тему из трёх вариантов: породы собак, напитки или грибы. ').lower()
    
    word = get_word(filename)
    ans = '_' * len(word)
    print('У Вас есть %d попыток, чтобы угадать слово из' % COUNT_END, len(word), 'букв.')
    
    game_end = False
    while not game_end:
        letter = input('Введите букву: ').lower()
        check_result = check_input(letter, dict_letters)
        if check_result == 1:
            print('Вы ввели некорректный символ.')
        elif check_result == 2:
            print('Вы уже вводили такую букву.')
        else:
            dict_letters.remove(letter)
            ans, count_attempt = apply_letter(letter, word, ans, count_attempt)
            print(ans, '\n')
            if count_attempt == 0:
                game_end = True
                print('У Вас закончились все попытки. К сожалению, Вы проиграли. Было загадано слово %s.' % word)
            else:
                if ans.find('_') == -1:
                    game_end = True
                    print('Ура! Вы победили!')
```

```python
if __name__ == "__main__":
    main()          
```
