```python
def get_set(filename):
    wfile = open(filename, 'r', encoding='utf-8')
    words = set() 
    for st in wfile.read().lower().split():
        if st.isalpha():
            words.add(st)
    return words
```

```python
set_br = get_set('Бродский.txt')
set_ahm = get_set('Ахматова.txt')
from pprint import pprint as pp
print('Пересечение множеств:')
pp(set_br & set_ahm)
print('Симметрическая разность множеств:')
pp(set_br ^ set_ahm)
```
